```python
import mysql.connector
from mysql.connector import Error

# Function to create a database connection
def create_connection():
    connection = None
    try:
        connection = mysql.connector.connect(
            host='Chamanchi',  
            user='mysql.infoschema',  
            password='venu',  
            database='sqlprogms'  
        )
        print("Connection to MySQL DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")
    return connection

# Function to create a registration
def create_registration(connection, name, email, date_of_birth):
    cursor = connection.cursor()
    try:
        query = "INSERT INTO Registration (Name, Email, DateOfBirth) VALUES (%s, %s, %s)"
        cursor.execute(query, (name, email, date_of_birth))
        connection.commit()
        print("Registration created successfully")
    except Error as e:
        print(f"The error '{e}' occurred")

# Function to read registrations
def read_registrations(connection):
    cursor = connection.cursor()
    cursor.execute("SELECT * FROM Registration")
    return cursor.fetchall()

# Function to update a registration
def update_registration(connection, registration_id, name, email, date_of_birth):
    cursor = connection.cursor()
    try:
        query = "UPDATE Registration SET Name = %s, Email = %s, DateOfBirth = %s WHERE ID = %s"
        cursor.execute(query, (name, email, date_of_birth, registration_id))
        connection.commit()
        print("Registration updated successfully")
    except Error as e:
        print(f"The error '{e}' occurred")

# Function to delete a registration
def delete_registration(connection, registration_id):
    cursor = connection.cursor()
    try:
        query = "DELETE FROM Registration WHERE ID = %s"
        cursor.execute(query, (registration_id,))
        connection.commit()
        print("Registration deleted successfully")
    except Error as e:
        print(f"The error '{e}' occurred")

# Example Usage
connection = create_connection()

# Create a new registration
create_registration(connection, "venu", "venu@gmail.com", "2003-21-10")

# Read all registrations
registrations = read_registrations(connection)
print(registrations)

# Update a registration
update_registration(connection, 1, "balaji", "balaji@gmail.com", "2002-01-02")

# Delete a registration
delete_registration(connection, 1)

# Close the connection
if connection.is_connected():
    connection.close()
    print("MySQL connection is closed")
```
